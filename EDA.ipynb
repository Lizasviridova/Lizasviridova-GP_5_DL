```python
df_new = df.drop(columns=['id', 'Unnamed: 0'])
```

```python
df_new 
```

```python
df_new['date'] = pd.to_datetime(df_new['date'])
df_new['month'] = df_new['date'].dt.month
df_new['year'] = df_new['date'].dt.year
```

```python
Вернемся к пропускам. Заполним их средним значением по этим столбцам, так как их относительно мало, около 6% и 4%, что не исказит наш анализ
```

```python
mean_bedrooms = df['bedrooms'].mean()
mean_bathrooms = df['bathrooms'].mean()
```

```python
df_new['bedrooms'].fillna(mean_bedrooms, inplace=True)
df_new['bathrooms'].fillna(mean_bathrooms, inplace=True)
```

```python
В датасете были обнаружены аномалии: дома с 0 спален и 0 ванных, а также одна запись с 33 спальнями

азменим информацию по дому с 33 спальнями, на медианное значение, так как 33 спальни, нам ничем не помогут
```

```python
median_bedrooms = df_new['bedrooms'].median()
df_new.loc[df_new['bedrooms'] == 33, 'bedrooms'] = median_bedrooms
```

```python
df_new['bedrooms'].max()
```

```python
Для домов с 0 спален и 0 ванных также можно использовать медиану
```

```python
df_new.loc[df_new['bedrooms'] == 0, 'bedrooms'] = median_bedrooms
df_new.loc[df_new['bathrooms'] == 0, 'bathrooms'] = df_new['bathrooms'].median()
```

```python
добавим новые признаки:

- Цена за квадратный фут жилой площади
- Категория реновации
- Возраст дома на момент продажи
```

```python
df_new['price_per_sqft'] = df_new['price'] / df_new['sqft_living']
```

```python
df_new['renovated'] = df_new['yr_renovated'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df_new['house_age'] = df_new['yr_built'].apply(lambda x: 2026 - x)
```
